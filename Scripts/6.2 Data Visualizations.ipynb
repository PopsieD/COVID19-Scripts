```python
#correlation heat map
corr= df_small.corr()
```

```python
corr
```

```python
sns.heatmap(corr)
```

A positive coefficeint mean when one variable increases, so will the other. With a negative coefficient,as one variable increases, the other will decrease. With a range from 0 to 1, a 0 coefficient indicates no relationship, 0.1-0.3 repesents a weak relationship, 0.3-0.5 indicates a moderate relationship, and 0.5-1.0 respresents a strong relationship between the variables.

```python
#scatterplot: Total Population and Total Households
sns.scatterplot(x= 'Total_Population' , y='Total_Households' , data= df_small)
```

The variables have a strong possitive correlation.

```python
#scatterplot: Total Households and Avg Income
sns.scatterplot(x= 'Average_Income(50k-100k)' , y= 'Total_Households' , data= df_small)
```

The variables have a strong positive correlation, possibly skewed by the end data point.

```python
#scatterplot: White Only and Avg Income
sns.scatterplot(x= 'Average_Income(50k-100k)' , y='White_only' , data= df_small)
```

These two variables would most likely have a smaller correlation with out the end data point which skews the results.

```python
#Pair Plot
sns.pairplot(df_small)
```

The pairplot shows each scatterplot relating to the correlation coefficient in the correlation heat map.

Total_Households and Average_Income:Total_Households is strongly correlated to Average_Income

Asian_only and High_Income:  There is a strong positive correlation that is intiguing implying the more asian citizens in a county, the more people will have a higher income

Black_only and Low_Income: The highest correlation with the Black_only variable is Low_Income

White_only and Average_Income: The highest correlation with Average_Income is White_only
